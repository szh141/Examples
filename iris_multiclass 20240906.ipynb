```python
import pandas as pd
from sklearn import datasets
import numpy as np

iris = datasets.load_iris() # returns dictionary

print(type(iris))

data = iris['data']
target = iris['target']
print(iris['feature_names'])
print(iris['target'])

test = np.concatenate((data,target.reshape(-1,1)),
                      axis=1)

"""
concatenate needs two arrays of same dimension

print(target.shape)
print(target.reshape(-1,1).shape)
output:
(150,)
(150, 1)

reshape(-1,1) calculate the first dimension
"""

df = pd.DataFrame(test,
                     columns= iris['feature_names'] + ['target'])

df
```

```python
# preprocessing

from sklearn.preprocessing import LabelEncoder

# Handle missing values if necessary
df = df.dropna()

# Separate input features and target variable
X = df.drop(columns=['target'])
y = df['target'].values

# Encode the target variable y (Flower)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)  # Converts the three flower types into 0, 1, or 2

# Check the number of features
n_features = X.shape[1]
```

```python
from sklearn.model_selection import train_test_split

# Split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Define the model
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(n_features,)))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3 neurons for 3 classes (Iris-setosa, Iris-versicolor, Iris-virginica)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=150, batch_size=32, verbose=0)
```

```python
# Evaluate the model
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print('Test Accuracy: %.3f' % acc)
```

```python
# Example prediction
import numpy as np

sample = np.array([[5.0, 3.6, 1.4, 0.2]])  # Example input for a new flower
prediction = model.predict(sample)

# Convert the prediction probabilities into a class label
predicted_class = np.argmax(prediction, axis=1)
print('Predicted class:', predicted_class)
```
