```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a Synthetic Dataset
# Let's create a simple synthetic dataset of linear and quadratic data
x = np.linspace(0, 10, 100)
y_linear = 2 * x + 3
y_quadratic = x**2 - 5 * x + 10

# Step 2: Store the Dataset in an HDF5 File

"""
with h5py.File('synthetic_data.h5', 'w') as f:
    f.create_dataset('X', data=x)
    f.create_dataset('Linear', data=y_linear)
    f.create_dataset('Quadratic', data=y_quadratic)
"""

f3 = h5py.File('synthetic_data.h5', 'w')
dataset1 = f3.create_dataset('X', data=x)
dataset1 = f3.create_dataset('Linear', data=y_linear)
dataset1 = f3.create_dataset('Quadratic', data=y_quadratic)
f3.close()

# Step 3: Read the Data Back from the HDF5 File
f2 = h5py.File('synthetic_data.h5', 'r')
x_read = f2['X'][:]
y_linear_read = f2['Linear'][:]
y_quadratic_read = f2['Quadratic'][:]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x_read, y_linear_read, label='Linear')
plt.title('Linear Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_read, y_quadratic_read, label='Quadratic', color='red')
plt.title('Quadratic Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.tight_layout()
plt.show()
```
