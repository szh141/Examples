np.random.randn(): To generate a random number/numbers from a standard normal distribution. A standard normal distribution will have a mean of zero and standard deviation of 1.

```python
'''Generate 10000 random numbers from a standard normal distribution'''
x=np.random.randn(10000)

'''plot the generated standard normal random numbers'''
sns.kdeplot(x,fill=True,alpha=0.2,common_norm=True)
plt.axvline(x=x.std(),linestyle='-.',color='k') # A vertical line to show standard deviation
plt.axvline(np.median(x),linestyle='-.',color='green',label = 'median') # A vertical line to show standard deviationplt.axvline(x=x.mean(),linestyle='-.',color='orange') # A vertical line to mean deviation
plt.legend()
plt.grid()
plt.show()
```

np.random.normal(): To generate a random number/numbers from a normal distribution with desired mean and standard deviation. Let's generate 10000 normally distributed data with a mean of 5 and standard deviation of 8, then plot it.

```python
'''10000 random normal numbers with mean=5 and std=8'''
x=np.random.normal(5,8,10000) #(mean , std , size)

'''plot the generated numbers'''
sns.kdeplot(x,fill=True,alpha=0.1,color='green')
plt.grid()
plt.axvline(x=x.std(),linestyle='-.',color='b',label='std')
plt.axvline(x=x.mean(),linestyle='-.',color='k',label='mean')
plt.legend()
```

np.random.poisson(): To generate a random number/numbers from a poisson distribution with specified mean or variance.

Poisson distribution, random discrete

https://www.youtube.com/watch?v=5uLxFmhYnl0


```python
'''generate 50 random numbers from a poisson distribution with a mean/variance =0.2'''
r = np.random.poisson(0.2, size=50) # lambda = mean or variance

'''plot the generated numbers'''
sns.kdeplot(r,fill=True,alpha=.1,color='g')
plt.axvline(x=r.mean(),ls="--",color='blue',label='mean')
plt.axvline(x=np.var(r),ls="--",color='k',label='var')
plt.axvline(x=np.median(r),ls="--",color='orange',label='median')
plt.grid()
plt.legend()
```

let’ increase lambda (mean/variance) and see whether it will approximate to a normal distribution

```python
'''generate 100 random numbers from a poisson distribution with a mean/variance =10'''
r = np.random.poisson(10, size=100) # lambda = mean or variance

'''plot the generated distribution'''
sns.kdeplot(r,fill=True,alpha=.1,color='g')
plt.axvline(x=r.mean(),ls="--",color='blue',label='mean')
plt.axvline(x=np.var(r),ls="--",color='k',label='var')
plt.axvline(x=np.median(r),ls="--",color='orange',label='median')
plt.grid()
plt.legend()
```
