```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data: clusters with noise
np.random.seed(42)  # For reproducibility
cluster_1 = np.random.normal(loc=(5, 5), scale=0.5, size=(100, 2))
cluster_2 = np.random.normal(loc=(10, 10), scale=1.0, size=(150, 2))
noise = np.random.uniform(low=0, high=15, size=(50, 2))

# Combine into a single dataset
data = np.vstack([cluster_1, cluster_2, noise])

plt.scatter(data[:,0],data[:,1])
```

```python
from sklearn.cluster import DBSCAN

# Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=10)
labels = dbscan.fit_predict(data)

plt.scatter(data[:,0],data[:,1],c=labels, s=24)
#print(f"Estimated number of clusters: {n_clusters_}")
```

The silhouette score ranges from -1 (incorrect clustering) to +1 (highly dense clustering), with scores around zero indicating overlapping clusters.

```python
from sklearn.metrics import silhouette_score

# Silhouette Score
score = silhouette_score(data, labels)
print(f"Silhouette Score: {score}")
```
