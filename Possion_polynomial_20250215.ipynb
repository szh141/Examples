# Poisson

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parameters
lambda_value = 20  # Average calls per hour
k_values = np.arange(0, 41, 1)  # Number of calls to consider

# Calculate Poisson PMF
pmf_values = poisson.pmf(k_values, lambda_value)

# Plotting the distribution
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, color='skyblue', edgecolor='black')
plt.axvline(x=25, color='red', linestyle='--', label='k = 25 (Peak)')
plt.title('Poisson Distribution: Call Volume Prediction')
plt.xlabel('Number of Calls (k)')
plt.ylabel('Probability')
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# Probability of exactly 25 calls
prob_25_calls = poisson.pmf(25, lambda_value)
print(f"Probability of receiving exactly 25 calls in an hour: {prob_25_calls:.4f}")

# Cumulative probability for <= 25 calls
cum_prob_up_to_25 = poisson.cdf(25, lambda_value)
print(f"Probability of receiving 25 or fewer calls: {cum_prob_up_to_25:.4f}")
```

```python
# root mean square error

import numpy as np

# Actual vs Predicted values
actual = np.array([3, -0.5, 2, 7])
predicted = np.array([2.5, 0.0, 2, 8])

# Step 1: Calculate squared errors
squared_errors = (actual - predicted) ** 2

# Step 2: Find the mean of squared errors
mse = np.mean(squared_errors)

# Step 3: Take the square root to get RMSE
rmse = np.sqrt(mse)

print("Root Mean Squared Error:", rmse)
```

# feature engineering, polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# Example data
X = np.array([[2], [3], [4]])

# Polynomial transformation
degree = 2
poly = PolynomialFeatures(degree)
X_poly = poly.fit_transform(X)

print("Original Features:", X)
print("Polynomial Features:", X_poly)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import numpy as np

# Generate synthetic data
X = np.arange(1, 10).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25, 36, 49, 64, 81])

# Polynomial transformation
degree = 2
poly = PolynomialFeatures(degree)
X_poly = poly.fit_transform(X)

# Fit a linear model
model = LinearRegression()
model.fit(X, y) # linear
y_pred1 = model.predict(X)
model.fit(X_poly, y)# polynomial
y_pred2 = model.predict(X_poly)

# Plot results
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred1, color='red', label='Polynomial Fit')
plt.legend()
plt.show()

plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred2, color='red', label='Polynomial Fit')
plt.legend()
plt.show()
```
