```python
import pandas as pd
# Create a sample DataFrame
data = {'full_name': ['Michael John son', 'Sarah Williams', 'David Br own', 'Emily Davis', 'Jessica Wilson']}
df = pd.DataFrame(data)
df
```

```python
df[['first_name', 'last_name']] = df['full_name'].str.split(' ', expand=True, n=1)
df
```
