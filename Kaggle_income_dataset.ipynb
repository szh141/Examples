```python
#Drop the duplicate columns
df.drop(['educational-num'], axis = 1, inplace = True)
df.shape
```

```python
#Unique value of the columns
df.nunique()
```

```python
df['age'].value_counts()
```

```python
plt.figure(figsize = (15, 6))
df['age'].value_counts().plot.bar(legend=True)
plt.show()
```

```python
colors = ["#ff5733","#ffe033", "#d7ff33", "#36ff33", "#33ecff", "#3355ff", "#f933ff", "#ff3390", "#c6c1e0"]
df['education'].value_counts().plot.bar(color=colors,legend=True)
```

```python
#Univariate analysis explores each variable in a data set, separately.

plt.figure(figsize = (18, 6))

sns.histplot(data = df['education'],kde=True)
plt.show()
```

```python
# Bivariate analysis: Analyse two columns age and income
sns.histplot(data = df, kde = True, x = 'age', hue = 'income')
```

```python
# Get aggregations of multiple columns for each target income group
age_and_hours_columns= ["age", "hours-per-week"]

df.groupby(["income"])[age_and_hours_columns].agg(['mean', 'std', 'max', 'min'])
```
