```python
import numpy as np
from matplotlib import pyplot as plt

x1 = [np.arange(0,10+1)]
y1 = [-i*(i-14) for i in x1]
print(x1, y1)

x2 = [np.arange(0,5+1)]
y2 = [3*i for i in x2]
print(x2, y2)

x3 = [np.arange(5,10+1)]
y3 = [15+(i-5)**2 for i in x3]
print(x3, y3)

plt.plot(x1, y1, x2, y2, x3, y3)  # numpy array, cannot plot
plt.show()
```

```python
# create dataset 3
x3 = [*range(5,10+1)]
y3 = [15+(i-5)**2 for i in x3]
print(x3, y3)

# create dataset 2
x2 = [*range(0,5+1)]
y2 = [3*i for i in x2]
print(x2, y2)

# create dataset 1
x1 = [*range(0,10+1)]
y1 = [-i*(i-14) for i in x1]
print(x1, y1)
```

```python
plt.plot(x1, y1, x2, y2, x3, y3)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'sr:', label='A')
plt.legend()
plt.ylabel('y')
plt.title('quadratic')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, '-.g^', label='B')
plt.legend()
plt.title('linier')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'o-b', label='C')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x1, y1, 'sr:')
plt.plot(x2, y2, '-.g^')
plt.plot(x3, y3, 'o-b')
plt.xlabel('x')
plt.xticks([0, 3, 4, 6, 8, 9, 10])
plt.text(5, 30, "area")

plt.show()
```
