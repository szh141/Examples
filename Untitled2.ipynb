```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

population = np.random.rand(500) # generating sequence of 500 random numbers
sample_size = [10, 20 ,40,100] # list of increasing sample size

fig, ax = plt.subplots(1,4 ,figsize=(20,4.5)) # create figure object

# loop through sample size and subplots & plot the distribution
for  ax ,sample_size in zip(ax.flatten(),sample_size):
    random_sample = np.random.choice(population, sample_size)# choose random samples
    sns.kdeplot(population, label= 'populaton distribution',ax=ax) #kde plot of population
    sns.kdeplot(random_sample, label = 'sample distribution',ax=ax)# kde plot of samples
    ax.axvline(x=population.mean(),linestyle='-.',color='blue',label='population_mean')
    ax.axvline(x=random_sample.mean(),linestyle='-.',color='orange',label='sample_mean')
    ax.legend()
    ax.set_title(f'Sample size = {sample_size}')
```

```python
import string

alphabet = list(string.ascii_lowercase) # list of 26 aplhabets. just to use as sample names
population = np.random.normal(10,2,10000) # random normal sequence of 10000 variables

mean_of_samples = [] # empty list to store mean of each of 26 samples
for i in alphabet: # loo through the list to choose 26 samples
    i = np.random.choice(population, 70) # choose sample of 70 variables
    mean_of_samples.append(i.mean()) # add mean of each sample to the list


sns.kdeplot(population, label='distribution of population')
sns.kdeplot(mean_of_samples, label='distribution of sample means')

plt.axvline(x=population.mean(),linestyle='-.',label='population mean')
plt.axvline(x=np.array(mean_of_samples).mean(),linestyle='-.',label='sample mean',color='orange')
plt.legend()
```
