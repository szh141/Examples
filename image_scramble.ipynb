```python
from skimage import data
from matplotlib import pyplot as plt
import numpy as np

import random
from datetime import datetime

img = data.camera()
temp = img.flatten()
img_s = np.zeros(img.shape)

a1 = list(temp)
a1



for i in range(0,(img.shape)[0]):
   for j in range(0,(img.shape)[1]):
       img_s[i,j] = int(a1.pop(np.random.randint(0,len(a1))))
```

```python
plt.subplot(221).imshow(img)
plt.subplot(222).hist(img)
plt.subplot(223).imshow(img_s)
plt.subplot(224).hist(img_s)
plt.show()
```

```python
plt.subplot(221).imshow(img)
plt.subplot(222).hist(list(img.flatten()))
plt.subplot(223).imshow(img_s)
plt.subplot(224).hist(list(img_s.flatten()))
plt.show()

```
