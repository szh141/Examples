```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the dark theme and a custom color palette
sns.set_theme(style="dark", palette="muted")

# Example plot
data = sns.load_dataset("penguins")
sns.histplot(data=data, x="flipper_length_mm", hue="species")
plt.show()
```

```python
import seaborn as sns

# Load the dataset
iris = sns.load_dataset("iris")

# Plot pairwise relationships
sns.pairplot(iris, hue="species")
plt.show()
```

```python
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# Create a linear model plot
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# Create a categorical plot
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.show()
```

```python
import seaborn as sns

# Load the dataset
penguins = sns.load_dataset("penguins")

# Create a grid of plots
g = sns.FacetGrid(penguins, col="species")
g.map(sns.histplot, "flipper_length_mm")
plt.show()
```

```python
import seaborn as sns

# Load the dataset
iris = sns.load_dataset("iris")

# Create a distribution plot
sns.histplot(iris['petal_length'], kde=True)
plt.show()
```

```python
import seaborn as sns

# Load the dataset
iris = sns.load_dataset("iris")

# Create a joint plot
sns.jointplot(x="petal_length", y="petal_width", data=iris, kind="scatter")
plt.show()
```

```python
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# Create a violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```
