#df.loc[ ] vs df.iloc[ ]

label vs integer

```python
# Example: Retrieve data for a specific row and column
df.loc['Row_Label', 'Column_Label']

# Example: Slice rows and columns based on labels
df.loc['Start_Row':'End_Row', 'Start_Column':'End_Column']

# Example: Boolean indexing with .loc[]
df.loc[df['Column'] > 10]

# Example: Select specific rows and columns
df.loc[['Row1', 'Row2'], ['Column1', 'Column2']]

# Example: Set values for a specific row and column
df.loc['Row_Label', 'Column_Label'] = new_value
```

```python
# Example: Retrieve data for a specific row and column using integer positions
df.iloc[2, 3]

# Example: Slice rows and columns based on integer positions
df.iloc[1:4, 0:2]

# Example: Select specific rows and columns using integer positions
df.iloc[[0, 2], [1, 3]]

# Example: Boolean indexing with .iloc[]
df.iloc[df['Column'] > 10]
```

Difference between .iloc[ ] and .loc[ ]

.iloc[] uses integer positions for indexing, while .loc[] relies on labels.
.iloc[] is inclusive on both ends during slicing, whereas .loc[] is inclusive on the start and exclusive on the end.

```python
# can also do boolean indexing w/o .loc or .iloc

selected_rows = df[df['Age'] > 25]
```

.map() Function

The .map() function is primarily used for substituting each value in a Series with another value. It can be applied to a single column or an entire DataFrame.

```python
import pandas as pd

# Sample DataFrame
data = {'Category': ['A', 'B', 'C', 'A']}
df = pd.DataFrame(data)

# Define a mapping dictionary
category_mapping = {'A': 'Alpha', 'B': 'Beta', 'C': 'Charlie'}

# Map values using the dictionary
df['Category_Name'] = df['Category'].map(category_mapping)

df
```
