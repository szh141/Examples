```python
import pandas as pd

names=['Robert','Joseph','David','Kasie','Derick']
age=[23,34,21,55,49]
profession=['Software engineer','Data Analyst',
            'Data Scientist','Doctor','Actor']
salary=[60000,75000,85000,80000,200000]
df=pd.DataFrame({'Names':names,'Age':age,
                 'Profession':profession,
                 'Salary':salary
})

df
```

The rename() method has a parameter called columns. This parameter takes a dictionary object as an argument. We have passed the dictionary as an argument. In this dictionary, the key (Profession) is the current name of the column, and the value (Job_Title) is the name we want to replace it with. Since we want to modify the DataFrame, we set the inplace parameter to True.

```python
df.rename(columns={'Profession':'Job_Title'}, inplace = True)
df
```

```python
df.rename(columns={'Names':'First_Name','Salary':'Salary_Per_Year'}, inplace = True)
df
```

rearrange column order for readability

```python
desired_order=['First_Names','Job_Title','Salary_Per_Year','Age']
df = df[desired_order]
df
```

```python
#  MultiIndex of Columns
df.set_index(['First_Names','Age'],inplace = True)
df
```

```python
df.reset_index(inplace=True)
df
```

```python
columns_to_drop =['Age','Salary_Per_Year']
df.drop(columns=columns_to_drop,inplace=True)
df
```
