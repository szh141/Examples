```python
import pandas as pd

# Sample dataframe with U.S. cities etc.
df = pd.DataFrame({
    'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago', 'Los Angeles'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing', 'Clothing', 'Electronics'],
    'Sales': [100, 200, 150, 250, 300, 180, 220, 270],
    'Quantity': [2, 5, 3, 7, 6, 4, 8, 9]
})

# Grouping by 'City' and 'Category' and summing up the other columns
grouped = df.groupby(['City', 'Category']).sum()

print(grouped)

# Grouping by 'City' and 'Category' and calculating sum, mean, and max for Sales and Quantity
grouped = df.groupby(['City', 'Category']).agg({
    'Sales': ['sum', 'mean', 'max'],
    'Quantity': ['sum', 'mean', 'max']
})

print(grouped)
```

#Using `apply()` and `map()` for Data Transformation

`apply()` allows us to apply a function along the axis of a DataFrame or on a Series. `map()` is a Series method used to replace each value in the Series with another value. Below are examples of using these functions in Pandas:

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}

df = pd.DataFrame(data)

# Define a function to calculate tax (for example, 10% tax)
def calculate_tax(salary):
    return salary * 0.10

# Apply the function to the 'Salary' column
df['Tax'] = df['Salary'].apply(calculate_tax)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 78, 92]
}
df = pd.DataFrame(data)

# Define a mapping function to categorize scores
def categorize_score(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    else:
        return 'C'

# Use map() to apply the categorize_score function to the 'Score' column
df['Grade'] = df['Score'].map(categorize_score)

# Display the updated DataFrame
print(df)
```

Use the `query()` function when we need to perform filtering operations on a DataFrame and prefer concise, readable syntax. It is especially useful when the filtering conditions are complex or when constructing query strings dynamically based on user input or program logic.

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'X': range(5, 11),
    'Y': range(20, 140, 20),
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles', 'Chicago']
})

# Using query() to filter rows
filtered_df = df.query('(X < 8) & (City == "New York")')

print(filtered_df)
```
