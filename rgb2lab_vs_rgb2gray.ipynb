The CIELAB (Lab) colour space is particularly useful due to its ability to separate colour from luminosity, making it a powerful tool for tasks such as grayscale conversion, brightness adjustment, and creating natural-looking inverted images.

L for its brightness, a for its green-red tint, and b for its blue-yellow tint. The separation of intensity from colour makes it independent of any device and capable of capturing a vast range of shades.

```python
import numpy as np
from skimage import io, data
from skimage.color import rgb2lab, lab2rgb,rgb2gray
import matplotlib.pyplot as plt

im = data.chelsea()
im1 = rgb2lab(im)

# Set the color channel values (a and b) to zeros, effectively removing color information and leaving only luminosity:
im1[...,1] = im1[...,2] = 0
im1 = lab2rgb(im1)
```

rgb2gray is a weighted RGB value
Y = 0.2125 R + 0.7154 G + 0.0721 B
https://scikit-image.org/docs/stable/auto_examples/color_exposure/plot_rgb_to_gray.html

```python
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'), plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im1), plt.axis('off'), plt.title('LAB-Grayscale image', size=20)
plt.show()

im2 = rgb2gray(im)
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'), plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im2,cmap='gray'), plt.axis('off'), plt.title('rgb2gray-Grayscale image', size=20)
plt.show()
```
